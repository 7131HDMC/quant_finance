```python

import yfinance as yf
import plotly.express as px
```

Baixamos os dados das ações ordinaria e prefencias da Petrobras

```python
petr4 = yf.download('PETR4.SA', start='2020-09-09', end='2023-09-09', interval='1d')

petr3 = yf.download('PETR3.SA',start='2020-09-09', end='2023-09-09', interval='1d')
```

```python
ratio = petr4.Close/petr3.Close
ratio.head()
```

```python
fig = px.line(x=ratio.index, y=ratio)

fig.add_hline(y=ratio.mean(), line_width=5, line_color="green")
fig.add_hline(y=(ratio.mean()-ratio.std()), line_width=3,
              line_dash="dash", line_color="orange")
fig.add_hline(y=(ratio.mean()+ratio.std()), line_width=3,
              line_dash="dash", line_color="orange")
fig.add_hline(y=(ratio.mean()-2*ratio.std()), line_width=5,
              line_dash="dash", line_color="red")
fig.add_hline(y=(ratio.mean()+2*ratio.std()), line_width=5,
              line_dash="dash", line_color="red")

fig.update_layout(xaxis_rangeslider_visible=False, title_text='Razão entre PETR4 e PETR3',template = 'simple_white',width=800,height=500)

fig.show()
```
